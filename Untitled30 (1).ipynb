1, Loading and Preprocessing

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load the diabetes dataset
diabetes_data = datasets.load_diabetes()
```

```python
# Convert to DataFrame 
df = pd.DataFrame(data=diabetes_data.data, columns=diabetes_data.feature_names)
df['target'] = diabetes_data.target
```

```python
print(df.head())
```

```python
# Checking for missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0]) 
```

```python
# Filling missing values 
df.fillna(df.mean(), inplace=True)  
```

```python
# Separate features and target
X = df.drop('target', axis=1)
y = df['target']
```

```python
# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
# Convert to DataFrame 
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)
```

```python
print(X_normalized_df.head())
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized_df, y, test_size=0.2, random_state=42)
```

```python
print(f'Training data shape: {X_train.shape}, Testing data shape: {X_test.shape}')
```

3, Building the ANN model

```python
pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
```

ANN model

```python
# ANN model
model = Sequential()
model.add(layers.Input(shape=(X_train.shape[1],)))  #Input layer
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='linear'))

```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
```

```python
model.summary()
```

4, Training the ANN Model

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Evaluate the model on the test set
test_loss, test_mae = model.evaluate(X_test, y_test, verbose=2)
```

```python
print(f'Test Loss: {test_loss:.4f}, Test MAE: {test_mae:.4f}')
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

5, Evaluating the Model

```python
#  test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
```

```python
# Calculate R² Score
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Mean Squared Error (MSE): {mse:.4f}')
print(f'R² Score: {r2:.4f}')
```

6, Improving the model

```python
# Improved Model 
Improved_model = keras.Sequential([layers.Input(shape=(X_train.shape[1],)),  
    layers.Dense(128, activation='relu'), 
    layers.Dropout(0.2), 
    layers.Dense(64, activation='relu'),  
    layers.Dense(32, activation='relu'),  
    layers.Dense(1)])
```

```python
# Compile the Improved Model
Improved_model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
                       loss='mean_squared_error',
                       metrics=['mae'])
```

```python
# Train the Improved model
Improved_history= Improved_model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
Improved_test_loss, Improved_test_mae =Improved_model.evaluate(X_test, y_test, verbose=1)
Improved_y_pred = Improved_model.predict(X_test)
```

```python
Improved_mse = mean_squared_error(y_test,Improved_y_pred )
Improved_r2 = r2_score(y_test,Improved_y_pred)
```

```python
print(f'Improved Test Loss (MSE): {Improved_test_loss:.4f}')
print(f'Improved Test MAE: {Improved_test_mae:.4f}')
print(f'Improved Mean Squared Error (MSE): {Improved_mse:.4f}')
print(f'Improved R² Score: {Improved_r2 :.4f}')
```

Changes Made in  the Model:-

1,Increased Neurons- Neurons in  the first  hidden layer increased to 128 neurons.

2,Added a dropout layer with a rate of 0.2 after the first hidden layer to prevent over fitting.

3,Added second hidden layer with 64 neurons.

4,Added third hidden  layer with 32 neurons.

5, Set the learning rate to 0.001.
